```python
import pandas as pd
import seaborn as sns
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
covid_data = pd.read_csv('covid_data_2020MAR31.csv', sep=';')
```

```python
covid_data = covid_data.assign(date = pd.to_datetime(covid_data.date, format = "%Y-%m-%d"))
```

```python
covid_data.head()
```

```python
covid_data.recovered.sum()
covid_data.country.value_counts().sort_values(ascending =False).head(20)
```

```python
plt.figure(figsize=(12,7))
covid_data.country.value_counts().sort_values(ascending =False).head(20).plot(kind='barh')
```

```python
covid_data_20202MAR29 = (covid_data
                .query("date == datetime.date(2020, 3, 29)")
                .filter(['country', 'confirmed', 'dead', 'recovered'])
                 .groupby('country')
                 .agg('sum')
                 .sort_values('confirmed', ascending = False)
                 .reset_index()


)

covid_data_20202MAR29
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data = covid_data_20202MAR29
                ,x = 'confirmed'
                , y = 'dead')
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data = covid_data_20202MAR29
                ,x = 'confirmed'
                , y = 'dead', hue = 'recovered')
```

```python
confirmed_by_date_xchina = (covid_data

                    .query("country != 'China'")
                    .filter(['date', 'confirmed'])
                    .groupby('date')
                    .agg('sum')
                    .reset_index()
                        
                        
                        )
confirmed_by_date_xchina                       
```

```python
confirmed_by_date_xchina.plot(kind ='line')
```

```python
sns.set()
plt.figure(figsize=(12,7))
sns.lineplot(data= confirmed_by_date_xchina, 
                    x= 'date', y= 'confirmed')
```

```python
confirmed_by_date_china_xchina= (covid_data
                    .assign(china_flg = np.where(covid_data.country == 'China', 'China', 'Not China'))
                    .filter(['date', 'confirmed', 'china_flg'])
                    .groupby(['date', 'china_flg'])
                    .agg('sum')
                    .reset_index()
    )
confirmed_by_date_china_xchina
```

```python
# confirmed_by_date_china_xchina[confirmed_by_date_china_xchina['china_flg'] == 'China']
```

```python
sns.lineplot(data = confirmed_by_date_china_xchina,x ='date',
                 y = 'confirmed', hue = 'china_flg')
```

```python
confirmed_by_date_top_countries =(covid_data
                    .filter(['date', 'country', 'confirmed'])
                    .query('country in ["US", "China", "Nigeria", "Italy", "Spain"]')
                    .groupby(['date', 'country']) 
                    .agg('sum')
                    .reset_index()

)
confirmed_by_date_top_countries
```

```python
plt.figure(figsize =(12,8))
sns.lineplot(data =confirmed_by_date_top_countries, x = 'date', y = 'confirmed',
               hue = 'country' )
```

```python
confirmed_by_country_top15 = (covid_data
            .query('date == datetime.date(2020, 3, 29)')
            .filter(['confirmed', 'country'])
            .groupby('country')
            .agg('sum')
            .sort_values('confirmed', ascending =False)
            .reset_index()
            .iloc[:15,:]

)
confirmed_by_country_top15
```

```python
plt.figure(figsize =(15,10))
sns.barplot(data = confirmed_by_country_top15
            , y ='country'
            , x = 'confirmed'
            , color = 'darkred'
            )
plt.title('Countries with highest covid cases', fontsize = 20, fontweight = 'bold', color = 'magenta')
```

```python
covid_data.head()
```

```python
confirmed_by_country_top15 = (covid_data
            .query('country in ["Nigeria", "Ghana", "Senegal", "Angola"]')
            .filter(['date','confirmed', 'country', 'recovered'])
            .groupby('country')
            .agg('sum')
            .sort_values('confirmed', ascending =False)
            .reset_index()
            .iloc[:15,:]

)
confirmed_by_country_top15
```

```python
sns.barplot(data = confirmed_by_country_top15, x = 'confirmed', y ='country')
```
